```python
pattern_size = 4

def hasABBA(s):
    if len(s) != pattern_size:
        return False
    all_distinct = len(set(s[0:pattern_size/2])) == pattern_size/2
    symmetric = all(s[i] == s[pattern_size - i - 1] for i in range(0,pattern_size/2))
    return all_distinct and symmetric

def parse(s):
    inside = []
    outside = []
    current = ""
    is_inside = False
    for c in s:
        if c == '[':
            outside.append(current)
            is_inside = True
            current = ""
        elif c == ']':
            inside.append(current)
            is_inside = False
            current = ""
        else:
            current = current + c
    if s[-1] == ']':
        inside.append(current)
    else:
        outside.append(current)
    return (inside,outside)

def hasABBAsubstring(s):
    for i in range(0,len(s) - pattern_size + 1):
        if hasABBA(s[i:i+pattern_size]):
            return True
    return False

def supportsTLS(s):
    inside,outside = parse(s)
    return any(map(hasABBAsubstring, outside)) and all(map(lambda s: not hasABBAsubstring(s), inside))

assert supportsTLS("abba[mnop]qrst")
assert not supportsTLS("abcd[bddb]xyyx")
assert not supportsTLS("aaaa[qwer]tyui")
assert supportsTLS("ioxxoj[asdfgh]zxcvbn")

# part 1
with open("input7.txt", 'r') as file_h:
    count = 0
    for l in file_h.readlines():
        if supportsTLS(l):
            count += 1
    print count
    
def correspond(aba,bab):
    return aba[0] == bab[1] and aba[1] == bab[0]

def is_ABA_or_BAB(s):
    return s[0] == s[2]

def get_ABAs_or_BABs(s):
    result = []
    for i in range(0,len(s) - 3 + 1):
        if is_ABA_or_BAB(s[i:i+3]):
            result.append(s[i:i+3])
    return result

def supportsSSL(s):
    inside,outside = parse(s)
    abas = [aba for s in outside for aba in get_ABAs_or_BABs(s)]
    babs = [bab for s in inside for bab in get_ABAs_or_BABs(s) ]
    for aba in abas:
        for bab in babs:
            if correspond(aba, bab):
                return True
    return False

assert supportsSSL("aba[bab]xyz")
assert not supportsSSL("xyx[xyx]xyx")
assert supportsSSL("aaa[kek]eke")
assert supportsSSL("zazbz[bzb]cdb")

# part 2
with open("input7.txt", 'r') as file_h:
    count = 0
    for l in file_h.readlines():
        if supportsSSL(l):
            count += 1
    print count
```
